```python
import numpy as np
import pandas as pd
import scipy.stats as st
import random
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Maximum Likelihood

#### $$\text{Maximum Likelihood} = \sum_{x_i \\ \in \\ \Omega}\text{ln} \\ ( \\ \text{Pr} \\ ( \\ x_i \\ | \\ \mu, \\ \sigma \\ ) \\ )\$$

```python
data = st.norm(4, 3).rvs(1000)
```

```python
def max_like_mean(arr, mean):
    return np.log(st.norm.pdf(arr, mean, 1)).sum()
```

```python
def max_like_std(arr, std, mu):
    return np.log(st.norm.pdf(arr, mu, std)).sum()
```

```python
mean = np.linspace(-10, 11, 300)
ml = [max_like_mean(data, mu) for mu in mean]

mean_estimated = mean[ml.index(max(ml))]

plt.figure(figsize=(13, 6))

plt.plot(mean, ml, color='red');
plt.xticks(np.arange(-10, 12))
plt.axvline(x = mean_estimated, color='k', linestyle='--');
mean_estimated
```

```python
std = np.linspace(0.001, 11, 300)
ml = [max_like_std(data, sd, 4.18) for sd in std]

std_estimated = std[ml.index(max(ml))]

plt.figure(figsize=(13, 6))

plt.plot(std, ml, color='red');
plt.xticks(np.arange(0, 12))
plt.axvline(x = std_estimated, color='k', linestyle='--');
std_estimated
```
