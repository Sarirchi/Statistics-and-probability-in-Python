```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### Horizontal BarPlot

```python
plt.figure(dpi=80)

sns.barplot([100, 180, 50, 20, 120], ['C#', 'Python', 'R', 'MatLab', 'JS'], orient='h')

plt.show();
```

### Vertical BarPlot

```python
plt.figure(dpi=80)

sns.barplot(['C#', 'Python', 'R', 'MatLab', 'JS'], [100, 180, 50, 20, 120], orient='v')

plt.show();
```

### Dist Plot

```python
data = stats.norm(0, 1).rvs(1000)

plt.figure(dpi=80)

sns.displot(data, kind='kde')

plt.show();
```

### Histogram

```python
plt.figure(dpi=80)

sns.distplot(data)

plt.show();
```

### Stacked Horizontal BarPlot

```python
plt.figure(dpi=80)

df = pd.DataFrame({
    'Name': ['Ali', 'Sadegh', 'Mohamad'],
    'Apple': [60, 20, 30],
    'Banana': [20, 40, 20]
})


sns.barplot(x='Name', y='Apple', data=df, label="Apple", color="red", alpha=.5)
sns.barplot(x='Name', y='Banana', bottom=[60, 20, 30], data=df, label="Banana", color="orange", alpha=.5)

plt.legend()
plt.show()
```

### Stacked Vertical BarPlot

```python
plt.figure(dpi=80)

df = pd.DataFrame({
    'Name': ['Ali', 'Sadegh', 'Mohamad'],
    'Apple': [60, 20, 30],
    'Banana': [20, 40, 20]
})


sns.barplot(x='Apple', y='Name', data=df, label="Apple", color="red", alpha=.5, orient='h')
sns.barplot(x='Banana', y='Name', left=[60, 20, 30], data=df, label="Banana", color="orange", alpha=.5, orient='h')

plt.legend()
plt.show()
```

### Pie Chart

```python
plt.figure(dpi=80)

plt.pie([40, 10, 30, 20],
        labels=['Python', 'Skala', 'R', 'Julia'],
        colors=['#a3adf7', '#f2acd8', '#acf2d3', '#f2d6ac'],
        explode=[0.07, 0, 0, 0],
        autopct='%1.2f%%', pctdistance=0.65)


plt.show()
```

### Scatter Plot

```python
x = np.random.random(100) * 100
y = x ** 2 * np.random.random(100) * 10

size = np.arange(1, 101)

plt.figure(dpi=80)

sns.scatterplot(x, y, color='red', alpha=0.4, size=size)

plt.show()
```

### Line Chart

```python
x = ['Far', 'Ord', 'Kho', 'Tir', 'Mor', 'Sha']
y = [20, 15, 30, 25, 40, 30]

plt.figure(dpi=80)

sns.lineplot(x, y, color='purple', alpha=0.3)

plt.show()
```

### Heatmap

```python
data = np.array([
    [1, 2, 4, 5, 3],
    [2, 3, 5, 1, 1],
    [2, 1, 3, 2, 1],
    [3, 4, 3, 2, 1]
])

plt.figure(dpi=80)

sns.heatmap(np.corrcoef(data), cmap='mako', annot=True)

plt.show()
```

### BoxPlot

```python
data = [12, 14, 15, 13, 12, 15, 20, 30]

plt.figure(dpi=80, figsize=(12, 5))

sns.boxplot(data, color='pink')

plt.show()
```

### Multiple Box Plot

```python
smoker = ['Smoker' if np.random.rand() < 0.5 else 'Non-Smoker' for _ in range(1000)]
heart_attack = stats.norm(0, 1).rvs(1000)

sns.boxplot(smoker, heart_attack, color='yellow')

plt.show()
```

### Joint Plot

```python
x1 = np.random.uniform(-1, 1, 1000)
x2 = x1 ** 2 + np.random.normal(0, 1, 1000) / 10

sns.jointplot(x1, x2, kind='scatter')

plt.show()
```

```python
sns.jointplot(x1, x2, kind='kde')

plt.show()
```

```python
sns.jointplot(x1, x2, kind='hex')

plt.show()
```

### Swarm Plot

```python
smoker = ['Smoker' if np.random.rand() < 0.5 else 'Non-Smoker' for _ in range(1000)]
heart_attack = stats.norm(0, 1).rvs(1000)

sns.swarmplot(smoker, heart_attack, color='red')

plt.show()
```
